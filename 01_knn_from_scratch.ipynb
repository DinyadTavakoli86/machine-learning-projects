# _KNN Algorithm from Scratch_

_This notebook demonstrates a simple implementation of the K-Nearest Neighbors (KNN) algorithm from scratch using Python, without relying on machine learning libraries such as scikit-learn_.

```python
import numpy as np
from collections import Counter

# ---- 1. Distance Function ----
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# ---- 2. KNN Class ----
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        return np.array([self._predict_one(x) for x in X])

    def _predict_one(self, x):
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_labels = [self.y_train[i] for i in k_indices]
        return Counter(k_labels).most_common(1)[0][0]

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)

#model = KNN(k=7)
accuracy=[]
for i in range(1,10):
    model = KNN(k=i)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy.append([np.mean(predictions == y_test)])
#model.fit(X_train, y_train)
#predictions = model.predict(X_test)
d={f"Index{accuracy.index(max(accuracy))}":float(max(accuracy)[0])}
d
#accuracy = np.mean(predictions == y_test)
#print(f"Accuracy: {accuracy:.2f}")
```
